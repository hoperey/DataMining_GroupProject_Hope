```python
# Notebook: 1_extract_transform.ipynb
# Author: Hope Kimandi
# Dataset: bank-additional-full.xlsx

# Import necessary libraries
import pandas as pd
import os

dataset_path = r'C:\Users\lenovo\OneDrive\Desktop\KEndy\ASSIGNMENTS\DSA Assignments\DSA 2040\Data mining_group project_Hope\DATA\bank-additional-full.xlsx'
df = pd.read_excel(dataset_path)

df.head()
```

```python
# Step 1: Initial Data Inspection

# Check data types and missing values
df.info()

# Summary statistics
df.describe(include='all')

# Check for duplicates
duplicates = df.duplicated().sum()
print(f'Total duplicates in dataset: {duplicates}')
```

```python
# Step 2: Data Cleaning

# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
# For categorical columns, replace NaNs with 'unknown'
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna('unknown')

# For numerical columns, replace NaNs with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
# Step 3: Feature Engineering

# Example: Create a categorical feature based on 'duration'
df['duration_category'] = pd.cut(
    df['duration'],
    bins=[-1, 100, 300, 600, float('inf')],
    labels=['short','medium','long','very_long']
)

# Encode target column 'y' as binary
df['y_binary'] = df['y'].apply(lambda x: 1 if x=='yes' else 0)
```

```python
# Step 4: Save Transformed Dataset

# Path to transformed folder
transformed_folder = r'C:\Users\lenovo\OneDrive\Desktop\KEndy\ASSIGNMENTS\DSA Assignments\DSA 2040\Data mining_group project_Hope\DATA\transformed'

# Create folder if it doesn't exist
os.makedirs(transformed_folder, exist_ok=True)

# Path to save the cleaned Excel
transformed_path = os.path.join(transformed_folder, 'bank_additional_transformed.xlsx')

# Save the cleaned dataset
df.to_excel(transformed_path, index=False)

print('Data cleaning and transformation completed. File saved to transformed folder.')
```
